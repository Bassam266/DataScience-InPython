```python
#back to scatteplot
g = sns.FacetGrid(movies, row="Genre", col="YearOfRelease" , hue="Genre")
kws = dict(s=50, linewidth=0.5, edgecolor='black')
g = g.map(plt.scatter,"RottenTomatoesRatings","AudienceRatings",**kws)
```

```python
#controlling Axes and adding diagonals
g = sns.FacetGrid(movies, row="Genre", col="YearOfRelease" , hue="Genre")
kws = dict(s=50, linewidth=0.5, edgecolor='black')
g = g.map(plt.scatter,"RottenTomatoesRatings","AudienceRatings",**kws)
g.set(xlim=(0,100),ylim=(0,100))
for ax in g.axes.flat:
    ax.plot((0,100),(0,100),c='gray',ls='--')
```

```python
sns.set_style("darkgrid")
f , axes = plt.subplots(2,2,figsize=(15,15))
k1 = sns.kdeplot(movies.BudgetMillion, movies.AudienceRatings , ax=axes[0,0])
k2 = sns.kdeplot(movies.BudgetMillion, movies.RottenTomatoesRatings, ax=axes[0,1])
k1.set(xlim=(-20,160))
k2.set(xlim=(-20,160))
z= sns.boxplot(data= movies, x='Genre', y='RottenTomatoesRatings',ax=axes[1,0])
k1 = sns.kdeplot(movies.RottenTomatoesRatings, movies.AudienceRatings, shade=True, shade_lowest=False, cmap="Reds",ax=axes[1,1])
```

```python
#styling  dashboards 
```

```python
sns.set_style("dark",{'axes.facecolor': 'black'})
f , axes = plt.subplots(2,2,figsize=(15,15))
k1 = sns.kdeplot(movies.BudgetMillion, movies.AudienceRatings , ax=axes[0,0],\
                shade=True, shade_lowest=True,cmap='inferno')
k2 = sns.kdeplot(movies.BudgetMillion, movies.RottenTomatoesRatings, ax=axes[0,1],\
                shade=True, shade_lowest=True,cmap='inferno')
k1.set(xlim=(-20,160))
k2.set(xlim=(-20,160))
z= sns.boxplot(data= movies, x='Genre', y='RottenTomatoesRatings',ax=axes[1,0])
k3 = sns.kdeplot(movies.RottenTomatoesRatings, movies.AudienceRatings, shade=True, shade_lowest=False, cmap="Blues_r",ax=axes[1,1])
```
